```python
%display latex
C(C_f, c_v, u) = C_f + c_v * u
C
```

```python
hat_c_f = 20
delta_c_f = 25
hat_c_v = 1
delta_c_v = 1.2
u_hat = 58
delta_u = 10
u_a = u_hat + delta_u

C_b(u) = C(hat_c_f, hat_c_v, u)
C_a(u) = C(hat_c_f+delta_c_f, hat_c_v+delta_c_v, u)
```

```python
Budget_Curve = plot(C_b, (u, 0, 100), color="black", linestyle="-")
Actual_Curve = plot(C_a, (u, 0, 100), color="red", linestyle="-")




(Budget_Curve+Actual_Curve).show(ymin=0)
```

```python
import matplotlib as mp
from matplotlib import pyplot

x_label = text('Production Units', (50, -10), color='black', fontsize=15)
y_label = text('Cost', (-10, 60), color='black', fontsize=15, rotation=90)
frame = x_label + y_label


Budget_Point = point((u_hat, C_b(u_hat)), color="black", marker="X", size=100)
Budget_Label = text('$\hat{C} = \hat{C_f} + \hat{c_v}\hat{u}$', (80,C_b(80)-15), color='black', fontsize=10)
Budget = Budget_Curve+Budget_Point+Budget_Label


Actual_Label = text('$C = C_f + c_vu$', (25,C_a(30)+10), color='red', fontsize=10)
Actual_Point = point((u_a, C_a(u_a)), color="red", marker="X", size=100)
Actual = Actual_Curve+Actual_Point+Actual_Label

C_a_0(u) = C_a(u)-delta_c_f
C_b_1(u) = C_b(u)+C_a(u_hat)-C_b(u_hat)

arrow_a_0 = plot(C_a_0, (u, 0, 100), color='pink') \
            + arrow((u_hat, C_b(u_hat)), (u_hat, C_a_0(u_hat)), color='blue', arrowsize=2, thickness=0.1) \
        + text("$\Delta_v\hat{u}$", (u_hat-4, C_a(u_hat)-50), color='black', fontsize=12)

arrow_delta_c_f = arrow((u_hat, C_a_0(u_hat)), (u_hat, C_a(u_hat)), color='blue', arrowsize=2, thickness=0.1) \
        + text("$\Delta_f$", (u_hat-3, C_a(u_hat)-15), color='black', fontsize=12)

arrow_delta_u = plot(C_b_1, (u, 0, 100), color="grey", linestyle="-.") \
        + arrow((u_hat, C_a(u_hat)), (u_a, C_a(u_hat)+delta_u*hat_c_v), color='blue', arrowsize=2, thickness=0.1) \
        + text('$c_v\Delta_u$', (u_hat+delta_u/2, C_a(u_hat)), color='black', fontsize=12)

arrow_delta_delta = arrow((u_a, C_b_1(u_a)), (u_a, C_a(u_a)), color='blue', arrowsize=2, thickness=0.1) \
                    + text('$\Delta_u\Delta_v$', (u_a+5, C_a(u_a)-5), color='black', fontsize=12)
    
# arrow((0, C_b(0)), (C_b(0), C_b(0)), color='black', arrowsize=5, thickness=0.1)+text("$\hat{C_f}$", (3, 10), color='black', fontsize=10)


(Budget+Actual+arrow_delta_c_f+arrow_delta_u+arrow_a_0+arrow_delta_delta).show(ymin=0, ymax=210)
```
